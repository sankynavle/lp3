**Assignment 1: Implement Gradient Descent Algorithm to find the local minima of a function.**

```python
def g(s,gf,mi,lr):
    
   
    s1=[s]
    x=s1[-1]
    for i in range(mi):
        x=x-lr*gf(x)
        s1.append(x)
    return s1,x,lr
    
    
    
    
    
def gd_f(x):
    return 8*x
    
    
list1,r,lr=g(3,gd_f,1300,0.001)
print(r)
print(list1)
```
